## State-Year MBS Data by Gender

```python
df_mbs_sa3_gender = df_mbs_sa3_service_1_np_new[
    df_mbs_sa3_service_1_np_new["DemographicGroup"].isin(["Males", "Females"])
]
df_mbs_sa3_gender.shape
```

```python
df_mbs_sa3_gender_state = df_mbs_sa3_gender.groupby(
    by=["Year", "StateTerritory", "Service", "DemographicGroup"], as_index=False
).agg(
    {
        "No_of_patients": "sum",
        "No_of_services": "sum",
        "Total_mbs_paid_$": "sum",
        "Total_provider_fees_$": "sum",
        "ERP": "sum",
        "Patient_OOP_per_service": "mean",
    }
)
df_mbs_sa3_gender_state
```

```python
# Calculate additional variable to understand state level mbs datapoints
df_mbs_sa3_gender_state["%_People_had_service"] = (
    df_mbs_sa3_gender_state["No_of_patients"] / df_mbs_sa3_gender_state["ERP"] * 100
)
df_mbs_sa3_gender_state["Out_of_Pocket"] = (
    df_mbs_sa3_gender_state["Total_provider_fees_$"]
    - df_mbs_sa3_gender_state["Total_mbs_paid_$"]
)
df_mbs_sa3_gender_state["No_of_service_per_person"] = (
    df_mbs_sa3_gender_state["No_of_services"]
    / df_mbs_sa3_gender_state["No_of_patients"]
)
df_mbs_sa3_gender_state["OOP_per_person_annual"] = (
    df_mbs_sa3_gender_state["No_of_service_per_person"]
    * df_mbs_sa3_gender_state["Patient_OOP_per_service"]
)
df_mbs_sa3_gender_state["Total_Provider_pp_per_service"] = df_mbs_sa3_gender_state[
    "Total_provider_fees_$"
] / (
    df_mbs_sa3_gender_state["No_of_patients"]
    * df_mbs_sa3_gender_state["No_of_service_per_person"]
)
df_mbs_sa3_gender_state["MBS_Rebate_pp_per_service"] = df_mbs_sa3_gender_state[
    "Total_mbs_paid_$"
] / (
    df_mbs_sa3_gender_state["No_of_patients"]
    * df_mbs_sa3_gender_state["No_of_service_per_person"]
)

df_mbs_sa3_gender_state.head(10)
```

```python
# exporting MBS-Census-State-Year-Service Level 1 (No demographic Data)
df_mbs_sa3_gender_state.to_csv(
    os.path.join(
        path, "clean_datasets/cleaned_csv/mbs_census_state_year_service_1_gender.csv"
    )
)
```

## State-Year MBS Data by Age Groups

```python
df_mbs_sa3_service_1_np_new["DemographicGroup"].unique()
```

```python
df_mbs_sa3_age_group = df_mbs_sa3_service_1_np_new[
    df_mbs_sa3_service_1_np_new["DemographicGroup"].isin(
        ["0-24", "25-44", "45-64", "65+"]
    )
]
df_mbs_sa3_age_group.shape
```

```python
# Calculate additional variable to understand state level mbs datapoints
df_mbs_agegroup_state["%_People_had_service"] = (
    df_mbs_agegroup_state["No_of_patients"] / df_mbs_agegroup_state["ERP"] * 100
)
df_mbs_agegroup_state["Out_of_Pocket"] = (
    df_mbs_agegroup_state["Total_provider_fees_$"]
    - df_mbs_agegroup_state["Total_mbs_paid_$"]
)

df_mbs_agegroup_state["No_of_service_per_person"] = (
    df_mbs_agegroup_state["No_of_services"] / df_mbs_agegroup_state["No_of_patients"]
)
df_mbs_sa3_gender_state["OOP_per_person_annual"] = (
    df_mbs_sa3_gender_state["No_of_service_per_person"]
    * df_mbs_sa3_gender_state["Patient_OOP_per_service"]
)
df_mbs_agegroup_state["Total_Provider_pp_per_service"] = df_mbs_agegroup_state[
    "Total_provider_fees_$"
] / (
    df_mbs_agegroup_state["No_of_patients"]
    * df_mbs_agegroup_state["No_of_service_per_person"]
)
df_mbs_agegroup_state["MBS_Rebate_pp_per_service"] = df_mbs_agegroup_state[
    "Total_mbs_paid_$"
] / (
    df_mbs_agegroup_state["No_of_patients"]
    * df_mbs_agegroup_state["No_of_service_per_person"]
)

df_mbs_agegroup_state.head(10)
```

```python
# exporting MBS-Census-State-Year-Service Level 1 (No demographic Data)
df_mbs_agegroup_state.to_csv(
    os.path.join(
        path, "clean_datasets/cleaned_csv/mbs_census_state_year_service_1_agegroup.csv"
    )
)
```
