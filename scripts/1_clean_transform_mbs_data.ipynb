# Clean Transform MBS Subsidised Services & Benefits Data

The script loads MBS datasets 2013-19, 2019-21 and 2021-22 and transforms each set to have measure values as separate columns instead of rows. The each dataset is exported to csv file. 

```python
# import libraries
import pandas as pd
import numpy as np
import os
```

## 1.MBS 2021-2022 Data Cleaning & Tranformation

```python
# import the mbs data files and assign to a dataframe

# setup path to original dataset of mbs fileds
path = r'/Users/patel/Documents/CF-Data Anaylst Course/portfolio_projects/mbs_analysis/datasets/'

#import required columns
filter_cols = ["Worksheet","Year","StateTerritory","GeographicUnit","GeographicCode","GeographicAreaName","GeographicGroup", "ServiceLevel", "Service", "DemographicGroup","MeasureName","MeasureValue","Flag","FlagExplanation"]

df_mbs_2021 = pd.read_csv(os.path.join(path, 'original_datasets/mbs_data/phc-mbs-2021-2022/AIHW_MBS_2122_CSV.csv'), encoding='ISO-8859-1', usecols=filter_cols, index_col=None)
df_mbs_2021.head(3)
```

```python
df_mbs_2021.shape
```

```python
df_mbs_2021.head()
```

### 1.2 Extract and transform SA3 Worksheet Data

SA3 worksheet contains Medicare subsidsed servicees data by Statistical Area Level 3 (geograpghical area) 

```python
# check the worksheet names to find SA3 reference
df_mbs_2021['Worksheet'].unique()
```

```python
df_mbs_2021[df_mbs_2021['Worksheet']=='SA3'].count()
```

```python
# extract SA3 data into a new dataframe. Used the copy function so the original dataframe is not impacted.
df_mbs_2021_sa3 = df_mbs_2021.loc[df_mbs_2021['Worksheet']=='SA3'].copy()
df_mbs_2021_sa3.head()
```

```python
print(df_mbs_2021_sa3.index.name)
df_mbs_2021_sa3.shape
```

```python
df_mbs_2021_sa3.isna().sum()
```

```python
df_mbs_2021_sa3[df_mbs_2021_sa3['StateTerritory'].isna()]
```

State Terriory had 4,032 blanls because SA3 Group aggregages are not for a specific state or terriorty. They are showing aggregrates of each SA3 group by demographics and service.

SA3 groups:
004-01 = Major cities - higher SES
004-02 = Major cities - medium SES
004-03 = Major cities - lower SES
004-04 = Inner regional
004-05 = Outer regional
004-06 = Remote (incl. very remote)

```python
# to ensure pivot table does not skip the blank stateterritory, filling in the blank cells.
df_mbs_2021_sa3_fill_na = df_mbs_2021_sa3.copy()
df_mbs_2021_sa3_fill_na['StateTerritory'].fillna('National-SA3Group', inplace=True)
```

```python
print(df_mbs_2021_sa3_fill_na.isna().sum())
df_mbs_2021_sa3_fill_na.shape
```

```python
# Use pivot_table to reshape the DataFrame
df_mbs_2021_sa3_pivot = df_mbs_2021_sa3_fill_na.pivot_table(index=['Worksheet', 'Year', 'StateTerritory', 'GeographicUnit', 'GeographicCode', 
                                                           'GeographicAreaName', 'GeographicGroup', 'ServiceLevel', 'Service', 'DemographicGroup'],
                            columns='MeasureName', 
                            values='MeasureValue',
                            aggfunc='first')  # 'first' is used to pick the first value in case of duplicates

# Reset index to make the DataFrame flat
df_mbs_2021_sa3_pivot.reset_index(inplace=True)
df_mbs_2021_sa3_pivot.head(5)
```

```python
df_mbs_2021_sa3_pivot.shape
```

```python
df_mbs_2021_sa3_pivot.isna().sum()
```

```python
#removed the column index 'MeasureValue' to simply the table
df_mbs_2021_sa3_pivot.columns.name = None
df_mbs_2021_sa3_pivot.head()
```

```python
# check number of rows and columns as expected
df_mbs_2021_sa3_pivot.shape
```

```python
# counting distinct values. Checking for 'National-SA3Group'
df_mbs_2021_sa3_pivot['StateTerritory'].value_counts()
```

```python
df_mbs_2021_sa3_pivot.rename(columns={'Medicare benefits per 100 people': 'Medicare benefits per 100 people ($)',
                                       'Total Medicare benefits paid':'Total Medicare benefits paid ($)', 'Total provider fees':'Total provider fees ($)'}, inplace=True)
df_mbs_2021_sa3_pivot.columns
```

```python
df_mbs_2021_sa3_pivot.drop('Worksheet', axis=1,inplace=True)
```

```python
df_mbs_2021_sa3_pivot.columns
```

```python
df_mbs_2021_sa3_pivot.to_csv(os.path.join(path, 'clean_datasets/mbs_data/2021-22_phc_mbs.csv'))
```
